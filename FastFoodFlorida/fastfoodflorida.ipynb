### Restaurantes de comida rápida en Florida, por condados

```python
import numpy as np  
import pandas as pd 
import folium
```

```python
# Dataset de restaurantes de EEUU
dffastfood = pd.read_csv('FastFoodRestaurants.csv')
dffastfood
```

```python
# Tomamos los que están en Florida
dff=dffastfood[dffastfood['province']=='FL']
dff
```

```python
# Creamos mapa centrado en Florida
mymap = folium.Map(location=[28.2701, -81.5844], zoom_start=7)
mymap
```

```python
# Marcadores agrupados con datos desplegables

from folium import plugins

fastfood = plugins.MarkerCluster().add_to(mymap)

for i in range(len(dff)):
    
    html=f"""
        <h2> {dff.iloc[i]['name']}</h2>
        <p>City: {dff.iloc[i]['city']}</p>
        
        <p>Address: {dff.iloc[i]['address']}</p>
        
        </p>
        <p><a href={dff.iloc[i]['websites']}>Restaurant website</a></p>
        """
    iframe = folium.IFrame(html=html, width=170, height=180)
    popup = folium.Popup(iframe, max_width=2650)
    
    
    folium.Marker(
        location=[dff.iloc[i]['latitude'], dff.iloc[i]['longitude']],
        icon=None,
        popup=popup
    ).add_to(fastfood)

mymap
```

```python
# Compruebo qué valores de postalCode no me gustan:
poscod=list(dff['postalCode'])
poscod
```

```python
# Cambio el postalCode que no me gusta para poder cambiar la serie a int
dff['postalCode'][dff.index[dff['postalCode']=='32837-6636' ]]=32837
dff['postalCode'][dff.index[dff['postalCode']=='33801-2601' ]]=33801
dff['postalCode'][8481]
dff['postalCode'][9831]
```

```python
dff.info()
```

```python
# Cambio el tipo de la columna postalCode a int ahora que ya se puede
dff['postalCode']=dff['postalCode'].astype(int)
dff.info()
```

```python
# Cargo dataset con los códigos postales de cada condado de Florida
floridacodes = pd.read_excel('floridacodes2.xlsx')
floridacodes
```

```python
# Prueba
floridacodes['ZIP Code'][1][9:]
```

```python
# Creo serie 'Code' en el dataframe con el código postal limpio
floridacodes['Code']=floridacodes['ZIP Code'].agg(lambda x: x[9:])
floridacodes
```

```python
# Elimino la antigua serie del código postal
floridacodes.pop('ZIP Code')
floridacodes.head()
```

```python
floridacodes.info()
```

```python
# Cambio el tipo de la serie Code a int
floridacodes['Code']=floridacodes['Code'].astype(int)
floridacodes.info()
```

```python
dff.rename(columns={"postalCode": "Code"},inplace=True)
dff
```

```python
# Hago merge de los dos dataframes, por la columna Code ahora que ambas son int
df=pd.merge(dff,floridacodes,on='Code',how='inner')
df
```

```python
# Creo dataframe con el número total de restaurantes por condado
dffloridatotal=pd.DataFrame(df.groupby('County')[['address']].agg('size'))
dffloridatotal.reset_index(inplace=True)
dffloridatotal.rename(columns={0: "Total"},inplace=True)
dffloridatotal
```

```python
# Choropleth en el mapa en el que ya tengo los marcadores

florida= r'FL_County.geojson' 

folium.Choropleth(
    geo_data=florida,
    data=dffloridatotal,
    columns=['County', 'Total'],
    key_on='properties.NAME',
).add_to(mymap)

mymap
```
